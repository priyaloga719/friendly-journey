```python
import pandas as pd

# Define speed threshold for spike
mean_speed = df['speed'].mean()
std_speed = df['speed'].std()
start_thresh = mean_speed + 1.0 * std_speed
end_thresh = mean_speed + 0.5 * std_speed

# Detect intervals
cme_intervals = []
in_cme = False
start_time = None

for i in range(len(df)):
    current_time = df['timestamp'].iloc[i]
    current_speed = df['speed'].iloc[i]
    
    if not in_cme and current_speed > start_thresh:
        in_cme = True
        start_time = current_time
    
    elif in_cme and current_speed < end_thresh:
        end_time = current_time
        if (end_time - start_time).total_seconds() >= 3600:  # at least 1 hour
            cme_intervals.append((start_time, end_time))
        in_cme = False

# Show results
print("Detected CME intervals:")
for start, end in cme_intervals:
    print(f"Start: {start}   End: {end}")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_score, recall_score, f1_score
from sklearn.model_selection import train_test_split

# Step 1: Prepare input features and target
feature_cols = ['speed_drop', 'flux_change', 'density_spike', 'temp_anomaly', 'speed_gradient']
X = df[feature_cols].fillna(0)
y = df['label']

# Step 2: Split into train/test (80/20)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Step 3: Train the model
model = LogisticRegression()
model.fit(X_train, y_train)

# Step 4: Predict on test set
y_pred_prob = model.predict_proba(X_test)[:, 1]

# Step 5: Apply threshold to convert to binary classification
threshold = 0.5
y_pred = (y_pred_prob >= threshold).astype(int)

# Step 6: Evaluate
precision = precision_score(y_test, y_pred, zero_division=0)
recall = recall_score(y_test, y_pred, zero_division=0)
f1 = f1_score(y_test, y_pred, zero_division=0)

print(f" Precision: {precision:.2f}")
print(f" Recall:    {recall:.2f}")
print(f" F1 Score:  {f1:.2f}")
```
